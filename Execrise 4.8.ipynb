# Direction 6

```python
# Group by 'user_id' and calculate the mean product price for each user
user_spending = ords_prods_merge.groupby('user_id')['prices'].mean().reset_index()
```

```python
# Create the spending_flag column based on the mean product price using loc
user_spending['spending_flag'] = 'Low spender'
```

```python
user_spending.loc[user_spending['prices'] >= 10, 'spending_flag'] = 'High spender'
```

```python
# Merge the spending_flag back into the original DataFrame
ords_prods_merge = ords_prods_merge.merge(user_spending[['user_id', 'spending_flag']], on='user_id', how='left')
```

```python
user_spending['spending_flag']
```

```python
# Display the updated DataFrame
print(ords_prods_merge.head())
```

# Directions 7

```python
# Group by 'user_id' and calculate the median days since prior order for each user
user_order_frequency = ords_prods_merge.groupby('user_id')['days_since_prior_order'].median().reset_index()
```

```python
# Create the order_frequency_flag column and initialize it to 'Frequent customer'
user_order_frequency['order_frequency_flag'] = 'Frequent customer'
```

```python
# Use loc to update the flags based on the median values
user_order_frequency.loc[user_order_frequency['days_since_prior_order'] > 20, 'order_frequency_flag'] = 'Non-frequent customer'
user_order_frequency.loc[(user_order_frequency['days_since_prior_order'] > 10) & 
                         (user_order_frequency['days_since_prior_order'] <= 20), 'order_frequency_flag'] = 'Regular customer'
```

```python
user_order_frequency['order_frequency_flag']
```

```python
# Merge the order_frequency_flag back into the original DataFrame
ords_prods_merge = ords_prods_merge.merge(user_order_frequency[['user_id', 'order_frequency_flag']], on='user_id', how='left')
```

```python
# Display the updated DataFrame
print(ords_prods_merge.head())
```
